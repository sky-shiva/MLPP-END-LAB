```python
from tensorflow import keras
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train_f=X_train.reshape(X_train.shape[0],28*28)
X_test_f=X_test.reshape(X_test.shape[0],28*28)
```

```python
model = keras.Sequential(
    [
        keras.layers.Dense(128,activation='relu',input_shape=(28*28,)),
        keras.layers.Dense(10,activation='softmax')
    ]
)
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train_f,y_train,epochs=5)
```

```python
test_lost,test_acc=model.evaluate(X_test_f,y_test)
print("Accuracy is : ",test_acc*100)
```
